Компьютер сам загадывает и сам угадывает число.

Импортируем библиотеку, которая нам пригодится для генерации случайных чисел.

```python
import numpy as np  #Импортируем библиотеку, которая нам пригодится для генерации случайных чисел.
from numpy import random

number = np.random.randint(1, 101) # загадываем число

def game_core_v1(number: int = 1) -> int: #Вариант первый перебор
    """Рандомно угадываем число 

    Args:
        number (int, optional): Загаданное число. Defaults to 1.

    Returns:
        int: Число попыток
    """
    count = 0

    while True:
        count += 1
        predict_number = np.random.randint(1, 101)  # предполагаемое число
        if number == predict_number:
            break  # выход из цикла если угадали
    return count

def game_core_v2(number: int = 1) -> int: #Вариант второй, прибаление или убывание на 1 от вероятного
    """Сначала устанавливаем любое random число, а потом уменьшаем
    или увеличиваем его в зависимости от того, больше оно или меньше нужного.
       Функция принимает загаданное число и возвращает число попыток 
       
    Args:
        number (int, optional): Загаданное число. Defaults to 1.

    Returns:
        int: Число попыток
    """
    count = 0
    predict = np.random.randint(1, 101)
    
    while number != predict:
        count += 1
        if number > predict:
            predict += 1
        elif number < predict:
            predict -= 1

    return count

def game_core_v3(number: int = 1) -> int: #Вариант третий, изменение границ вычисления вероятного числа
    """Сначала устанавливаем любое random число, а потом изменям границы поиска в зависимости от того, больше оно или меньше нужного.
    Т.е. если загаданное число больше вероятного, то нижнию границу для генерации вероятного числа принимаем равной вероятному.
    Т.е. если загаданное число меньше вероятного, то верхнюю границу для генерации вероятного числа принимаем равной вероятному.
    Повторяем вычисление выроятного числа в рамках новых границ.
    
       Функция принимает загаданное число и возвращает число попыток 
       
    Args:
        number (int, optional): Загаданное число. Defaults to 1.

    Returns:
        int: Число попыток
    """
    limit_a = 1
    count = 0
    limit_b = 101
    predict = np.random.randint(1, 101)
    
    while number != predict:  
        count += 1
        predict = np.random.randint(limit_a, limit_b)
        if number > predict:
            limit_a = predict
                       
        elif number < predict:
            limit_b = predict
    return count

def game_core_v4(number: int = 1) -> int: #Вариант четвертый, взятие среднего между двух границ, границы меняются от вероятного числа
    #Очень долго работает
    """Сначала берем вероятное число как половину верней границы.
    Если загаданное число больше вероятного, то вероятное число определяем как нижнию границу.
    Если загаданное число меньше вероятного, то вероятное число определяем как верхнюю границу.
    Вероятное число определяем как половину суммы границ.
    
       Функция принимает загаданное число и возвращает число попыток 
       
    Args:
        number (int, optional): Загаданное число. Defaults to 1.

    Returns:
        int: Число попыток
    """
    count = 1
    mn, mx = 1, 100
    predict = mx / 2
    while number != predict:
        count+=1
 
        if number > predict: 
            mn = predict 
        
        elif number < predict: 
            mx = predict 
        
        predict = round((mx + mn) / 2)
    
    return count

def score_game(random_predict) -> int:
    """За какое количство попыток в среднем за 1000 подходов угадывает наш алгоритм

    Args:
        random_predict ([type]): функция угадывания

    Returns:
        int: среднее количество попыток
    """
    count_ls = []
    np.random.seed(1)  # фиксируем сид для воспроизводимости
    random_array = np.random.randint(1, 101, size=(10000))  # загадали список чисел

    for number in random_array:
        count_ls.append(random_predict(number))

    score = int(np.mean(count_ls))
    print(f"Наш алгоритм угадывает число в среднем за:{score} попыток")
    return score


if __name__ == "__main__":
    # RUN
    score_game(game_core_v4)
```
